**Question 2:** In how many different states do the customers live in?

```python
import csv
datafile = "./sample_data/customers.csv"
estados_unicos = set()

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')
    next(csvreader) # Saltamos el encabezado

    for row in csvreader:
        # 1. Usamos .strip() para sacar espacios invisibles
        estado = row[4].strip()

        # 2. Solo lo agregamos si NO está vacío
        if estado != "":
            estados_unicos.add(estado)

print(f"Ahora el número de estados reales es: {len(estados_unicos)}")
print("Los estados son:", sorted(list(estados_unicos)))
```

**Question 3** What is the state with most customers?

```python
import csv

datafile = "./sample_data/customers.csv"

# 1. Creamos un diccionario para llevar la cuenta
conteo_estados = {}

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')
    next(csvreader) # Saltamos el encabezado

    for row in csvreader:
        estado = row[4] # Extraemos el estado

        # 2. Lógica de conteo
        if estado in conteo_estados:
            conteo_estados[estado] += 1
        else:
            conteo_estados[estado] = 1

# 3. Encontrar el estado con el número más alto
# Usamos la función max() sobre el diccionario
estado_maximo = max(conteo_estados, key=conteo_estados.get)
cantidad_maxima = conteo_estados[estado_maximo]

print(f"El estado con más clientes es {estado_maximo} con {cantidad_maxima} clientes.")
```

**Question 4** What is the state with the least customers?

```python
import csv

datafile = "./sample_data/customers.csv"

# 1. Diccionario para contar clientes por cada estado
conteo_estados = {}

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')
    next(csvreader)  # Saltamos la fila de encabezado

    for row in csvreader:
        # Usamos .strip() para eliminar espacios invisibles y trabajamos con la columna State (índice 4)
        estado = row[4].strip()

        # Solo contamos si el campo no está vacío
        if estado != "":
            if estado in conteo_estados:
                conteo_estados[estado] += 1
            else:
                conteo_estados[estado] = 1

# 2. Encontramos cuál es la cantidad mínima de clientes registrada
cantidad_minima = min(conteo_estados.values())

# 3. Identificamos TODOS los estados que tienen esa cantidad mínima (buscamos el empate)
estados_con_menos_clientes = []
for estado, total in conteo_estados.items():
    if total == cantidad_minima:
        estados_con_menos_clientes.append(estado)

# 4. Ordenamos alfabéticamente para que coincida con las opciones del quiz
estados_con_menos_clientes.sort()

# Mostramos el resultado final
print(f"La cantidad mínima de clientes es: {cantidad_minima}")
print(f"Los estados con esa cantidad son: {', '.join(estados_con_menos_clientes)}")
```

**Question 5:** What is the most common last name?

```python
import csv

datafile = "./sample_data/customers.csv"

# 1. Creamos un diccionario para contar los apellidos
conteo_apellidos = {}

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')
    next(csvreader) # Saltamos el encabezado

    for row in csvreader:
        apellido = row[2] # El apellido está en el índice 2

        # 2. Contamos cada vez que aparece un apellido
        if apellido in conteo_apellidos:
            conteo_apellidos[apellido] += 1
        else:
            conteo_apellidos[apellido] = 1

# 3. Buscamos el que tenga el número más alto
apellido_comun = max(conteo_apellidos, key=conteo_apellidos.get)
cantidad = conteo_apellidos[apellido_comun]

print(f"El apellido más común es '{apellido_comun}' y aparece {cantidad} veces.")
```

**Question #1:** How many unique orders are in the orders.csv file?

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

**Question #3:** What is the highest number of items per order?

**Question #4:** What is the number of orders placed in October 2021?

**Question #5:** Which customer spent the most amount of money in 2021?

**Question #6:** Historically, what is the best month for sales?

```python
import csv
from os.path import exists

order_file = "./sample_data/orders.csv"
customer_file = "./sample_data/customers.csv"

# Diccionarios y conjuntos para el análisis
conteo_items_por_orden = {}
ordenes_octubre_2021 = set()
gastos_por_cliente_2021 = {}
ventas_por_mes_historico = {}
ordenes_ya_procesadas = set()

with open(order_file, 'r') as f:
    csvreader = csv.reader(f)
    next(csvreader) # Saltamos el encabezado

    for row in csvreader:
        # 1. Limpieza básica: Saltamos filas incompletas
        if len(row) < 4 or not row[1].strip():
            continue

        customer_id = row[0].strip()
        order_id = row[1].strip()
        fecha = row[2].strip()
        total_orden = float(row[3])

        # --- A. Cálculos para Items (Preguntas 2 y 3) ---
        conteo_items_por_orden[order_id] = conteo_items_por_orden.get(order_id, 0) + 1

        # --- B. Cálculos Financieros (Preguntas 1, 4, 5 y 6) ---
        if order_id not in ordenes_ya_procesadas:
            # Pregunta 4: Octubre 2021
            if "2021-10" in fecha:
                ordenes_octubre_2021.add(order_id)

            # Pregunta 5: Gasto por cliente en 2021 (solo si hay fecha válida)
            if "2021" in fecha:
                gastos_por_cliente_2021[customer_id] = gastos_por_cliente_2021.get(customer_id, 0) + total_orden

            # Pregunta 6: Mejor mes (solo si la fecha tiene el formato correcto)
            if len(fecha) >= 7:
                mes_anio = fecha[:7] # Extrae "YYYY-MM"
                ventas_por_mes_historico[mes_anio] = ventas_por_mes_historico.get(mes_anio, 0) + total_orden

            ordenes_ya_procesadas.add(order_id)

# --- PROCESAMIENTO DE RESULTADOS ---
cantidades = list(conteo_items_por_orden.values())
promedio_items = sum(cantidades) / len(cantidades)
max_items = max(cantidades)
total_ordenes_unicas = len(ordenes_ya_procesadas)
total_octubre = len(ordenes_octubre_2021)
id_top_cliente = max(gastos_por_cliente_2021, key=gastos_por_cliente_2021.get)
mejor_mes_nombre = max(ventas_por_mes_historico, key=ventas_por_mes_historico.get)

# --- BUSQUEDA DEL NOMBRE DEL CLIENTE ---
nombre_cliente = "Desconocido"
with open(customer_file, 'r') as f:
    csvreader = csv.reader(f)
    next(csvreader)
    for row in csvreader:
        if row[0] == id_top_cliente:
            nombre_cliente = f"{row[1]} {row[2]}"
            break

# --- IMPRESIÓN FINAL ---
print(f"1. Total de órdenes únicas: {total_ordenes_unicas}")
print(f"2. Promedio de ítems por orden: {promedio_items:.2f}")
print(f"3. Máximo de ítems por orden: {max_items}")
print(f"4. Órdenes en Octubre 2021: {total_octubre}")
print(f"5. Cliente que más gastó: {nombre_cliente} (ID {id_top_cliente}) con ${gastos_por_cliente_2021[id_top_cliente]:.2f}")
print(f"6. Mejor mes histórico: {mejor_mes_nombre} con ${ventas_por_mes_historico[mejor_mes_nombre]:.2f}")
```
